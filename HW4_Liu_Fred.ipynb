```python
#5
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
grade = []
name = []
for key, value in marks.items():
    grade.append(value)
    name.append(key)
grade = np.array(grade)
name = np.array(name)
plt.bar(name, grade)
plt.title('Student Grade in Bar Chart')
plt.show()

plt.pie(grade, labels = name)
plt.legend(bbox_to_anchor=(1,1))
plt.title('Student Grade in Pie Chart')
plt.show()
```
