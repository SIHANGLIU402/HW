```r
#Q3Choice2
amazon = read.csv('amazon_orders.csv')
#remove all dollar signs
amazon$Item.Subtotal = as.numeric(gsub("\\$", "", amazon$Item.Subtotal))
amazon$Item.Subtotal.Tax = as.numeric(gsub("\\$", "", amazon$Item.Subtotal.Tax))
amazon$Item.Total = as.numeric(gsub("\\$", "", amazon$Item.Total))
#Remove the first roll because this order was cancelled
amazon2 = amazon[-c(1),]
#replace all NaN cells to 0
amazon2[is.na(amazon2)] = 0
amazon2$Order.Date = as.Date(amazon2$Order.Date, format= "%m/%d/%y")
  
amazon2
```

```r
#total spending
sum(amazon2$Item.Total)
```

```r
#Gives minimum, maximum, median, mean of the following three sets of data
summary(amazon2$Item.Subtotal)
summary(amazon2$Item.Subtotal.Tax)
summary(amazon2$Item.Total)
```

```r
#gives standard deviation of total cost
sd(amazon2$Item.Total)
```

```r
x = sum(amazon2$Item.Subtotal.Tax)/sum(amazon2$Item.Total)
sprintf('%0.1f%%', x*100)
```

```r
#gives total monthly spending
Feb_cost = amazon2[c(1:22),]
Mar_cost = amazon2[c(23:35),]
sum(Feb_cost$Item.Total)
sum(Mar_cost$Item.Total)
```

```r
#group cost by date and plot a bar chart
cost_by_date = aggregate(amazon2$Item.Total, by=list(amazon2$Order.Date), sum)
barplot(cost_by_date[,2], names.arg=cost_by_date[,1])
```

```r
#graph a bar chart to show the amount paied to each Seller
cost_by_seller = aggregate(amazon2$Item.Subtotal, by=list(amazon2$Seller), sum)
pie(cost_by_seller$x, label=cost_by_seller[,1], main='Payment to Each Seller', radius=.5, col = cost_by_seller[,1])
legend(x=.7, y=1.08, legend = cost_by_seller[,1], fill=cost_by_seller[,1], cex=.5)
```

```r
#multipleline comparison of monthly spending
line1 = amazon2[c(1:22),]
line2 = amazon2[c(23:35),]

plot(line1$Item.Total, type='l', main='Monthly Cost Comparison', xlab = 'Days', ylab = 'Cost', col='blue')
lines(line2$Item.Total, type='l', col='green')
legend(16, 300, legend=c("Line 1", "Line 2"), col=c("blue", "green"), lty=1, cex=0.8)
```
